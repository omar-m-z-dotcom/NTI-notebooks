```python
import sympy as sym

x,y,z = sym.symbols("x y z")
exp = x**3 * y+ + y**3 +z
exp
```

```python
derivative1_x = sym.diff(exp,x)
derivative1_x
```

```python
derivative1_y = sym.diff(exp,y)
derivative1_y
```

```python
derivative1_x = sym.diff(exp,x,2)
derivative1_x
```

```python
derivative1_y = sym.diff(exp,y,2)
derivative1_y
```

```python
sym.diff(x**4,x,x,x)
```
