```python
import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia page
url = "https://ar.wikipedia.org/wiki/%D8%A7%D9%84%D9%83%D9%88%D9%86"

# Send a GET request to fetch the HTML content of the page
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")

# Find the <p> tags which contain the main text content
paragraphs = soup.find_all("p")

# Extract the text from the <p> tags
text = "\n".join([p.get_text() for p in paragraphs])

# Print the extracted text
print(text)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
from arabert.preprocess import ArabertPreprocessor

# Define the model name
model_name = "malmarjeh/t5-arabic-text-summarization"
# Instantiate the Arabert preprocessor
preprocessor = ArabertPreprocessor(model_name="")

# Load the tokenizer and model on the GPU
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Instantiate the pipeline with the model and tokenizer
summarization_pipeline = pipeline("text2text-generation", model=model, tokenizer=tokenizer)

# Define the input texts
text = preprocessor.preprocess(text)

# Generate the summary
result = summarization_pipeline(text,
                                 max_length=400,  # Increase max_length to desired value
                                pad_token_id=tokenizer.eos_token_id,
                                num_beams=3,
                                repetition_penalty=0.8,
                                length_penalty=0.8,
                                no_repeat_ngram_size=3)[0]['generated_text']

print(result)
```

```python
from transformers import BertTokenizer, AutoModelForSeq2SeqLM, pipeline
from arabert.preprocess import ArabertPreprocessor

model_name="/content/mbert2mbert-arabic-text-summarization"
preprocessor = ArabertPreprocessor(model_name="")

tokenizer = BertTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
pipeline = pipeline("text2text-generation",model=model,tokenizer=tokenizer)

text = "شهدت مدينة طرابلس، مساء أمس الأربعاء، احتجاجات شعبية وأعمال شغب لليوم الثالث على التوالي، وذلك بسبب تردي الوضع المعيشي والاقتصادي. واندلعت مواجهات عنيفة وعمليات كر وفر ما بين الجيش اللبناني والمحتجين استمرت لساعات، إثر محاولة فتح الطرقات المقطوعة، ما أدى إلى إصابة العشرات من الطرفين."
text = preprocessor.preprocess(text)

result = pipeline(text,
            pad_token_id=tokenizer.eos_token_id,
            num_beams=3,
            repetition_penalty=3.0,
            max_length=200,
            length_penalty=1.0,
            no_repeat_ngram_size = 3)[0]['generated_text']
result
```
