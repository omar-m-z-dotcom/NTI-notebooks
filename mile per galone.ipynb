```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Create a list of column names, matching the expected number of columns in the data
column_names = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model_year", "origin", "name"]

# Load the data using pandas.read_csv, specifying the column names and delimiter
data_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
df = pd.read_csv(data_url, sep='\s+', names=column_names,na_values="?")

# Print the first few rows of the DataFrame to verify it was loaded correctly
print(df.head())
```

```python
df = df.drop(axis=1, columns=["name"])
df
```

```python
# Assuming you have your DataFrame loaded as 'df'

# Create a MinMaxScaler object
scaler = MinMaxScaler(feature_range=(-1, 1))

# Normalize all columns (excluding potential non-numeric columns)
df_normalized = pd.DataFrame(scaler.fit_transform(df.select_dtypes(include=[np.number])), columns=df.columns)

# Print the normalized DataFrame
print(df_normalized)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Calculate correlation matrix
correlation_matrix = df_normalized.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
