```python
import numpy as np
from tensorflow import keras
```

```python
from keras.datasets import mnist

(x_train,y_train),(x_test,y_test) = mnist.load_data()

y_test
```

```python
# x_train_new, y_train_new = x_train[(y_train <= 1)], y_train[(y_train <= 1)]
x_train_new, y_train_new = x_train, y_train
```

```python
# x_test_new , y_test_new = x_test[(y_test <= 1)], y_test[((y_test <= 1))]
x_test_new , y_test_new = x_test, y_test
```

```python
x_train_new.shape
```

```python
# Reshape to a 2D array with 784 elements per sample (corrected syntax)
x_train_final = x_train_new.reshape(x_train_new.shape[0], 784)
x_train_final.shape
```

```python
x_test_final = x_test_new.reshape(x_test_new.shape[0], 784)
x_test_final.shape
```

```python
x_train_final = x_train_final.astype("float32") / 255.0
x_test_final = x_test_final.astype("float32") / 255.0
```

```python
print(y_train_new.shape)
y_test_new.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model  = LogisticRegression(solver = 'lbfgs')

model.fit(x_train_final, y_train_new)

# Make predictions on the test set
predictions = model.predict(x_test_final)

# Evaluate the model using the correct test set
score = accuracy_score(y_test_new, predictions)
print(score)
```

```python
from sklearn import metrics

cm = metrics.confusion_matrix(y_test_new, predictions)
print(cm)
```
