```python
import random

# Generate a random integer between 1 and 100 (inclusive)
random_numbers = [random.randint(0, 1000) for i in range(1000)]
even_count = 0
odd_count = 0
for n in random_numbers:
  if(n == 0):
    print(f"n is: {n}")
    print("terminating")
    break
  if(n%2 == 0):
    even_count+=1
  else:
    odd_count+=1

print(f"even nums: {even_count}, odd nums: {odd_count}")

```

```python
# Create a list of first names and last names
first_names = ["Alice", "Bob", "Charlie", "Diana", "Eve"]
last_names = ["Smith", "Jones", "Williams", "Brown", "Miller"]
full_names=[]
for first_name in first_names:
  for last_name in last_names:
    full_names.append(first_name + " " + last_name)

final = []
for full_name in full_names:
  final.append(full_name.replace(" ","_"))

print(final)
```

```python
import random

# Generate a random integer between 1 and 100 (inclusive)
random_numbers1 = [random.randint(0, 1000) for i in range(1000)]
print(f"max number: {max(random_numbers1)}")
# create a new list to store unique elements
unique_list = []
# iterate over the original list
for x in random_numbers1:
  # check if the element is not already in the unique list
  if x not in unique_list:
    # add the element to the unique list
    unique_list.append(x)
print(f"unique_list: {unique_list}")

random_numbers2 = [random.randint(0, 1000) for i in range(1000)]

has_common = False
for n1 in random_numbers1:
  for n2 in random_numbers2:
    if n1 == n2:
      has_common = True
print(has_common)
```

```python
number = 5
def factorial(num):
  if num == 1:
    return num
  else:
    return num * factorial(num-1)
print(factorial(number))
```
