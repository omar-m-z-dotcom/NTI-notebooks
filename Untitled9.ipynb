```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("/content/Sunspots.csv")
data.drop(columns=['Month'],inplace=True)
data
```

```python

window_size = 12

# Define a function to create the time series
def create_time_series(data, window_size,inc):
    inputs = []
    outputs = []
    i = 0
    while i <(len(data) - window_size):
        inputs.append(data['Sunspots'].iloc[i:i+window_size].values)
        outputs.append(data['Sunspots'].iloc[i+window_size])
        i+=inc
    return np.array(inputs), np.array(outputs)

inputs, outputs = create_time_series(data, window_size,12)

# Reshape the inputs and outputs to have one row per input sequence
# Each row contains window_size inputs followed by a single output
inputs = inputs.reshape(-1, window_size)
outputs = outputs.reshape(-1, 1)

# Combine inputs and outputs horizontally
time_series_data = np.hstack((inputs, outputs))

# Convert to DataFrame
time_series1 = pd.DataFrame(time_series_data, columns=[f'x{i}' for i in range(window_size)] + ['target'])

print(time_series1)

inputs, outputs = create_time_series(data, window_size,1)

# Reshape the inputs and outputs to have one row per input sequence
# Each row contains window_size inputs followed by a single output
inputs = inputs.reshape(-1, window_size)
outputs = outputs.reshape(-1, 1)

# Combine inputs and outputs horizontally
time_series_data = np.hstack((inputs, outputs))

# Convert to DataFrame
time_series2 = pd.DataFrame(time_series_data, columns=[f'x{i}' for i in range(window_size)] + ['target'])

print(time_series2)

```

#first way - simple RNN

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

X = time_series1.drop('target', axis=1).values
y = time_series1['target'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))


scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train.reshape(-1, X_train.shape[-1])).reshape(X_train.shape)
X_test_scaled = scaler.transform(X_test.reshape(-1, X_test.shape[-1])).reshape(X_test.shape)

y_scaler = MinMaxScaler()
y_train_scaled = y_scaler.fit_transform(y_train.reshape(-1, 1))
y_test_scaled = y_scaler.transform(y_test.reshape(-1, 1))

model = Sequential()
model.add(SimpleRNN(32, input_shape=(X_train_scaled.shape[1], 1)))
model.add(Dense(32, activation='tanh'))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(X_train_scaled, y_train_scaled, epochs=15, batch_size=1, verbose=1)

y_pred = model.predict(X_test_scaled)

sns.set_style('whitegrid')

plt.figure(figsize=(10, 6))
plt.plot(y_test_scaled, color='blue', label='Actual')
plt.plot(y_pred, color='red', label='Predicted')
plt.title('Actual vs Predicted Sunspots')
plt.xlabel('Time')
plt.ylabel('Sunspots')
plt.legend()
plt.show()
```
