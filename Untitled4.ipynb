```python
import tensorflow as tf
import numpy as np
from transformers import TFAutoModelForSequenceClassification, AutoTokenizer
from sklearn.model_selection import train_test_split
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.optimizers import Adam
import tensorflow_datasets as tfds

# Load the IMDb movie review dataset
dataset, info = tfds.load("imdb_reviews", with_info=True, as_supervised=True)

print(dataset)
print(info)

# Split the dataset into train and test sets
train_dataset, test_dataset = dataset["train"], dataset["test"]

print(train_dataset)
print(test_dataset)


# Extract texts and labels from the datasets
train_texts = []
train_labels = []
test_texts = []
test_labels = []

for text, label in train_dataset:
    train_texts.append(text.numpy().decode("utf-8"))
    train_labels.append(label.numpy())
    print(train_texts)
    print(train_labels)

for text, label in test_dataset:
    test_texts.append(text.numpy().decode("utf-8"))
    test_labels.append(label.numpy())
    print(test_texts)
    print(test_labels)

# Convert labels to numpy arrays
train_labels = np.array(train_labels)
test_labels = np.array(test_labels)

print(train_labels)
print(test_labels)

# Load pre-trained tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = TFAutoModelForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)

# Tokenize input texts
X_train_tokenized = tokenizer(train_texts, padding=True, truncation=True, return_tensors="tf")
X_test_tokenized = tokenizer(test_texts, padding=True, truncation=True, return_tensors="tf")

print(X_train_tokenized)
print(X_test_tokenized)

# Compile the model
model.compile(optimizer=Adam(), loss=SparseCategoricalCrossentropy(from_logits=True), metrics=["accuracy"])

# Train the model
history = model.fit(
    X_train_tokenized,
    train_labels,
    validation_data=(X_test_tokenized, test_labels),
    batch_size=32,
    epochs=200,
)

# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test_tokenized, test_labels)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```
