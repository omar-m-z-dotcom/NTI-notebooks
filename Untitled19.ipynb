```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from prophet import Prophet
import warnings
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
import itertools

warnings.filterwarnings("ignore")
```

# loading the data

```python
df = pd.read_csv("MarketPricePrediction.csv")
df["month"] = pd.Categorical(
    df["month"],
    [
        "January",
        "February",
        "March",
        "April",
        "May",
        "June",
        "July",
        "August",
        "September",
        "October",
        "November",
        "December",
    ],
)
df.sort_values(by=["year", "month"], inplace=True)
df.reset_index(drop=True, inplace=True)


# convert months to numbers
def month_to_num(month):
    return {
        "January": 1,
        "February": 2,
        "March": 3,
        "April": 4,
        "May": 5,
        "June": 6,
        "July": 7,
        "August": 8,
        "September": 9,
        "October": 10,
        "November": 11,
        "December": 12,
    }[month]


df["month"] = df["month"].apply(month_to_num)
df
```

# data cleaning

```python
print(f"data description:\n{df.describe()}")
print("\ndata information:")
print(f"{df.info()}")
print(f"number of duplicate rows: {df.duplicated().sum()}")
```

```python
corr = df[["quantity", "priceMin", "priceMax", "priceMod"]].corr()
print(corr)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

```python
# plot the quantity, priceMin, priceMax, priceMod for each market
markets = df["market"].unique()
columns = ["quantity", "priceMin", "priceMax", "priceMod"]

fig, axs = plt.subplots(
    len(markets), len(columns), figsize=(20, 1 * len(markets) * len(columns))
)

for i, market in enumerate(markets):
    market_df = df[df["market"] == market]
    if market_df.shape[0] < 2:
        continue
    market_datetime = pd.to_datetime(
        market_df["year"].astype(str) + "-" + market_df["month"].astype(str)
    )
    market_df["datetime"] = market_datetime
    market_df.drop(
        columns=["year", "month", "date", "state", "city", "market"], inplace=True
    )
    # reorder columns
    market_df = market_df[["datetime", "quantity", "priceMin", "priceMax", "priceMod"]]
    for j, column in enumerate(columns):
        market_df_y = market_df[["datetime", column]]
        market_df_y.rename(columns={"datetime": "ds", column: "y"}, inplace=True)
        market_df_y.reset_index(drop=True, inplace=True)
        m = Prophet()
        m.fit(market_df_y)
        future = m.make_future_dataframe(periods=0, freq="M")
        forecast = m.predict(future)
        market_df_y["trend"] = forecast["trend"]
        market_df_y.set_index("ds", inplace=True)
        market_df_y.rename(columns={"y": column}, inplace=True)
        axs[i, j].plot(market_df_y[column], label=column)
        axs[i, j].plot(market_df_y["trend"], label=f"{column}_trend", linestyle="--")
        axs[i, j].set_title(f"{market} - {column}")
        axs[i, j].grid(True)
        axs[i, j].set_xlabel("Date")
        axs[i, j].legend()

plt.tight_layout()
plt.show()
```

# ADF analysis

```python
markets = df["market"].unique()
columns = ["quantity", "priceMin", "priceMax", "priceMod"]

fig, axs = plt.subplots(len(markets), len(columns), figsize=(20, 1*len(markets)*len(columns)))
usedlags = {'quantity': [], 'priceMin': [], 'priceMax': [], 'priceMod': []}
for i, market in enumerate(markets):
    market_df = df[df["market"] == market]
    if market_df.shape[0] < 2:
        continue
    market_datetime = pd.to_datetime(
        market_df["year"].astype(str) + "-" + market_df["month"].astype(str)
    )
    market_df["datetime"] = market_datetime
    market_df.drop(
        columns=["year", "month", "date", "state", "city", "market"], inplace=True
    )
    # reorder columns
    market_df = market_df[["datetime", "quantity", "priceMin", "priceMax", "priceMod"]]
    market_df.set_index("datetime", inplace=True)
    for j, column in enumerate(columns):
        series = market_df[column]
        diff = series.diff().dropna()

        # Perform ADF test on original series
        try:
            result = adfuller(series.dropna())
        except:
            continue
        print(f'{market} - {column} - ADF Statistic: {result[0]} - p-value: {result[1]} - usedlag: {result[2]}')

        # Perform ADF test on differenced series
        try:
            result = adfuller(diff)
        except:
            continue
        print(f'{market} - {column} (differenced) - ADF Statistic: {result[0]} - p-value: {result[1]} - usedlag: {result[2]}')
        usedlags[column].append(result[2])

        # Plot original series and differenced series
        axs[i,j].plot(series, label=f'{column}')
        axs[i,j].plot(diff, label=f'{column} (differenced)', linestyle='--')
        axs[i,j].set_title(f'{market} - {column}')
        axs[i,j].grid(True)
        axs[i,j].set_xlabel('Date')
        axs[i,j].legend()

plt.tight_layout()
plt.show()
for column in columns:
    print(f'{column} - mean usedlag: {round(np.mean(usedlags[column]))}')
    print(f'{column} - most common usedlag: {max(set(usedlags[column]), key=usedlags[column].count)}')
```

# ADF analysis 12 months

```python
markets = df["market"].unique()
columns = ["quantity", "priceMin", "priceMax", "priceMod"]

fig, axs = plt.subplots(len(markets), len(columns), figsize=(20, 1*len(markets)*len(columns)))
usedlags = {'quantity': [], 'priceMin': [], 'priceMax': [], 'priceMod': []}

for i, market in enumerate(markets):
    market_df = df[df["market"] == market]
    if market_df.shape[0] < 12:  # Check if market has at least 12 months of data
        continue
    market_datetime = pd.to_datetime(
        market_df["year"].astype(str) + "-" + market_df["month"].astype(str)
    )
    market_df["datetime"] = market_datetime
    market_df.drop(
        columns=["year", "month", "date", "state", "city", "market"], inplace=True
    )
    # reorder columns
    market_df = market_df[["datetime", "quantity", "priceMin", "priceMax", "priceMod"]]
    market_df = market_df.sort_values('datetime').iloc[:12]  # Select first 12 months of data
    market_df.set_index("datetime", inplace=True)
    for j, column in enumerate(columns):
        series = market_df[column]
        diff = series.diff().dropna()

        # Perform ADF test on original series
        result = adfuller(series.dropna())
        print(f'{market} - {column} - ADF Statistic: {result[0]} - p-value: {result[1]} - usedlag: {result[2]}')

        # Perform ADF test on differenced series
        result = adfuller(diff)
        print(f'{market} - {column} (differenced) - ADF Statistic: {result[0]} - p-value: {result[1]} - usedlag: {result[2]}')
        usedlags[column].append(result[2])

        # Plot original series and differenced series
        axs[i,j].plot(series, label=f'{column}')
        axs[i,j].plot(diff, label=f'{column} (differenced)', linestyle='--')
        axs[i,j].set_title(f'{market} - {column}')
        axs[i,j].grid(True)
        axs[i,j].set_xlabel('Date')
        axs[i,j].legend()

plt.tight_layout()
plt.show()
for column in columns:
    print(f'{column} - mean usedlag: {round(np.mean(usedlags[column]))}')
    print(f'{column} - most common usedlag: {max(set(usedlags[column]), key=usedlags[column].count)}')
```

```python
market_dfs = {}
# create a dataframe for each market
for market in df_copy["market"].unique():
    market_df = df_copy[df_copy["market"] == market]
    market_df = market_df.drop(columns=["market"])
    market_df["ds"] = pd.to_datetime(
        market_df["year"].astype(str) + "-" + market_df["month"].astype(str)
    )
    market_df.set_index("ds", inplace=True)
    market_df.drop(columns=["year", "month",'state','city'], inplace=True)
    market_dfs[market] = market_df
```

# testing SARIMA on priceMin column

```python
SARIMAX_priceMin_models = {}
SARIMAX_priceMin_results = {}
for market, _ in market_dfs.items():
    SARIMAX_priceMin_models[market] = None
    SARIMAX_priceMin_results[market] = None
# split the data into training and test data for each market and apply the model on the quantity column
progress = 0

for market, market_df in market_dfs.items():

    # split the data into training and test data in the ratio 80:20

    train = market_df["priceMin"].iloc[: int(len(market_df) * 0.8)]
    test = market_df["priceMin"].iloc[int(len(market_df) * 0.8) :]

    # apply the pipeline

    # Define the p, d, and q parameters to take any value between 0 and 2

    p = d = q = range(0, 2, 1)


    # Generate all different combinations of p, q and q triplets

    pdq = list(itertools.product(p, d, q))


    # Generate all different combinations of seasonal p, q and q triplets

    seasonal_pdq = [(x[0], x[1], x[2], x[3]) for x in list(itertools.product(p, d, q, range(0, 13, 1)))]


    # specify to use the Akaike Information Criterion (AIC) to select the best model

    best_aic = np.inf

    best_pdq = None

    best_seasonal_pdq = None

    temp_model = None


    for param in pdq:

        for param_seasonal in seasonal_pdq:


            try:

                temp_model = SARIMAX(train, order=param, seasonal_order=param_seasonal)

                results = temp_model.fit()


                # print(

                #     "SARIMAX{}x{}12 - AIC:{}".format(param, param_seasonal, results.aic)

                # )

                if results.aic < best_aic:

                    best_aic = results.aic

                    best_pdq = param

                    best_seasonal_pdq = param_seasonal

            except:

                # print("Unexpected error:", sys.exc_info()[0])

                continue
    print(

        "market {} Best SARIMAX{}x{} model - AIC:{}".format(
            market, best_pdq, best_seasonal_pdq, best_aic
        )
    )


    # Fit the best model

    model = SARIMAX(train, order=best_pdq, seasonal_order=best_seasonal_pdq)

    model_fit = model.fit(disp=False)


    # make prediction

    yhat = model_fit.predict(
        len(train),

        len(train) + len(test) - 1,
    )


    print(f"Predicted: {yhat.values}, Actual: {test.values}")
    mae = np.mean(np.abs(yhat.values - test.values))
    mse = np.mean((yhat.values - test.values) ** 2)
    rmse = np.sqrt(mse)
    print(f"MAE: {mae}, MSE: {mse}, RMSE: {rmse}")
    SARIMAX_priceMin_models[market] = model_fit
    SARIMAX_priceMin_results[market] = {
        "MAE": mae,
        "MSE": mse,
        "RMSE": rmse,
    }
    progress += 1
    print(f"Progress: {progress}/{len(market_dfs)}")

# overall results
overall_MAE = np.mean([result["MAE"] for result in SARIMAX_priceMin_results.values()])
overall_MSE = np.mean([result["MSE"] for result in SARIMAX_priceMin_results.values()])
overall_RMSE = np.mean([result["RMSE"] for result in SARIMAX_priceMin_results.values()])
print(
    f"Overall MAE: {overall_MAE}, Overall MSE: {overall_MSE}, Overall RMSE: {overall_RMSE}"
)
```
