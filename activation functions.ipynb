```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    sigmoid_x = sigmoid(x)
    return sigmoid_x * (1 - sigmoid_x)

# Generate x values
x = np.linspace(-10, 10, 100)

# Calculate sigmoid and its derivative values
sigmoid_values = sigmoid(x)
sigmoid_derivative_values = sigmoid_derivative(x)

# Plot sigmoid function and its derivative
plt.figure(figsize=(8, 6))
plt.plot(x, sigmoid_values, label='Sigmoid', color='blue')
plt.plot(x, sigmoid_derivative_values, label='Sigmoid Derivative', color='red', linestyle='--')
plt.title('Sigmoid Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('Function Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

# Generate x values
x = np.linspace(-10, 10, 100)

# Calculate tanh and its derivative values
tanh_values = tanh(x)
tanh_derivative_values = tanh_derivative(x)

# Plot tanh function and its derivative
plt.figure(figsize=(8, 6))
plt.plot(x, tanh_values, label='tanh', color='blue')
plt.plot(x, tanh_derivative_values, label='tanh Derivative', color='red', linestyle='--')
plt.title('Hyperbolic Tangent Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('Function Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# Generate x values
x = np.linspace(-10, 10, 100)

# Calculate ReLU and its derivative values
relu_values = relu(x)
relu_derivative_values = relu_derivative(x)

# Plot ReLU function and its derivative
plt.figure(figsize=(8, 6))
plt.plot(x, relu_values, label='ReLU', color='blue')
plt.plot(x, relu_derivative_values, label='ReLU Derivative', color='red', linestyle='--')
plt.title('ReLU Function and Its Derivative')
plt.xlabel('x')
plt.ylabel('Function Value')
plt.legend()
plt.grid(True)
plt.show()
```
